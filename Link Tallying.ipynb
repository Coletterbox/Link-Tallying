```python
import re
from collections import Counter
import matplotlib.pyplot as plt
```

```python
inputText = 'FB Marketplace Data.txt'

# with open(inputText, encoding='cp850') as f:
# with open(inputText, encoding='utf8') as f:
with open(inputText, errors="ignore") as f:
    lines = f.readlines()
```

```python
type(lines)
len(lines)

linkArray = []

for line in lines:
    linkArray.append(line.split('www.'))
```

```python
urlArray = []

with open(inputText) as file:
        for line in file:
            url = re.findall('https?://(?:[-\w.]|(?:%[\da-fA-F]{2}))+', line)
            if len(url) > 0:
                urlArray.append(url)
                
urlArray[:3]
```

```python
def flatten_list(urls):
    flat_list = []
    for element in urls:
        if type(element) is list:
            for item in element:
                flat_list.append(item)
        else:
            flat_list.append(element)
    return flat_list
```

```python
newUrlArray = flatten_list(urlArray)
```

```python
newUrlArray[:10]
```

```python
trimmedUrls = []

for url in newUrlArray:
    newUrl = url.split('%')
    trimmedUrls.append(newUrl[0])
    
print(trimmedUrls[:10])
```

```python
urls0 = []

for url in trimmedUrls:
    if '.' in url:
        urls0.append(url)
        
print(urls0[:10])
```

```python
urls1 = []

for url in urls0:
    if ('facebook' not in url) & ('fbcdn' not in url) & ('fb.me' not in url) & ('bit.ly' not in url) & ('instagram' not in url) & ('kfc' not in url) & ('landbot' not in url) & ('jotform' not in url) & ('zoopla' not in url) & ('landmarkshops' not in url) & ('ncl' not in url) & ('tatacliq' not in url) & ('nupad' not in url):
        urls1.append(url)
        
print(urls1)
```

```python
urls2 = []

for url in urls1:
    urls2.append(url.replace('https://', '').replace('www.', '').replace('http://', ''))
```

```python
Counter(urls2)
```

```python
w = Counter(urls2)

temp2 = plt.figure(figsize=(20, 5)) 
temp = plt.xticks(rotation=90)

plt.bar(w.keys(), w.values())

plt.suptitle('Frequency of Companies Advertising this Month')
plt.xlabel('Website')
plt.ylabel('Frequency')
plt.rcParams['font.size'] = 18

plt.show()
```
